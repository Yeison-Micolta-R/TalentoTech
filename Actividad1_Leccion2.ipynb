```python
#carga de paquetes y funciones para generar gráficos

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#generamos datos aleatorios para ser analizados

rango = np.random.RandomState(0) #crea un generador de números aleatorios utilizando la función RandomState de NumPy
x = np.linspace(0,10,500) #crea un generador de números aleatorios utilizando la función RandomState de NumPy
y = np.cumsum(rango.randn(500,6),0) #se genera una matriz de datos y de dimensiones 500x6. rango.randn(500,6)
```

#Graficar los datos

```python
plt.plot(x,y)
plt.legend('ABCDEF',ncol=2, loc='upper left')
```

#Cambio de estilo usando la funcion set de seaborn

```python
sns.set() #Carga esilos de seaborn
plt.plot(x,y)
plt.legend('ABCDEF',ncol=2, loc='upper left')
```

#Generando puntos de una función

```python
x_val = np.linspace(0,10,20)
y_val = x_val**2
plt.scatter(x_val,y_val)
```

#Agregando títulos a la gráfica y a los ejes

```python
plt.scatter(x_val,y_val)
plt.title('Desplazamiento') #Título del gráfico
plt.xlabel('Tiempo [Segundos]') # Etiqueta del eje X
plt.ylabel('Distancia [Metros]') # Etiqueta del eje Y
```

#Modificar los rótulos (puntos) de la gráfica

```python
plt.scatter(x_val,y_val, marker='*') # marker cambia los rótulos
plt.title('Desplazamiento') #Título del gráfico
plt.xlabel('Tiempo [Segundos]') # Etiqueta del eje X
plt.ylabel('Distancia [Metros]') # Etiqueta del eje Y
```

#Cambiando color y tamaño a los puntos

```python
plt.scatter(x_val,y_val, marker=',', color='g',s=150) # el atributo color modifica el color de los puntos (r=rojo,g=verde,b=azul)
# el atributo 's' modifica el tamaño de los puntos
plt.title('Desplazamiento') #Título del gráfico
plt.xlabel('Tiempo [Segundos]') # Etiqueta del eje X
plt.ylabel('Distancia [Metros]') # Etiqueta del eje Y
```

#HISTOGRAMAS

```python
import pandas as pd
data = np.random.multivariate_normal([0,0],[[5,2],[2,2]], size=2000)
data = pd.DataFrame(data,columns =['x','y'])

```

```python
for col in 'xy':
  plt.hist(data[col], alpha=0.5)
```

```python
for col in 'xy':
  sns.kdeplot(data[col],shade = True)
```

```python
sns.distplot(data['x'])
sns.distplot(data['y'])
```
