```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path ='/content/drive/MyDrive/Universidad/Cursos/Python Data Science/titanic.csv'
```

```python
df_titanic = pd.read_csv(path)
df_titanic
```

```python
C_M =df_titanic.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(C_M, annot=True, cmap="RdYlGn", cbar=True)
plt.title("Matriz de CorrelaciÃ³n")
plt.show()
```
