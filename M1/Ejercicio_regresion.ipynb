```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,2,2,3,4,4,5,6]
y=[2,3,4,4,4,6,5,7]
```

```python
x = np.array(x)
y = np.array(y)

xy = x*y
xy
```

```python
x2 = x**2
x2
```

```python
sumax= np.sum(x)
sumay= np.sum(y)
sumaxy= np.sum(xy)
sumax2= np.sum(x2)

sumax,sumay,sumaxy,sumax2
```

```python
m = ((8*sumaxy)-(sumax*sumay))/((8*sumax2)-sumax**2)
m
```

```python
promy = sumay/8
promx = sumax/8
promx, promy
```

```python
b = promy-(m*promx)
b
```

```python
plt.scatter(x,y)
plt.plot(x, m*x + b)
plt.show()
```
