```python
import pandas as pd
import matplotlib as plt
import numpy as np
```

```python
val_name=['Yeison', 'Alfredo']
val_h=[1.76,1.59] #m
val_w=[76, 75] #kg
val_bmi=[25,28]

data_dict = { "name":val_name,"height":val_h,"weigth":val_w,"BMI":val_bmi}

df = pd.DataFrame(data_dict)

df
```

```python
df["BMI"] = df["weigth"]/df["height"]**2
df
```

```python
df['Normal Range'] = pd.Series([])
```

```python
df['Normal Range'] = np.where((df['BMI']>=18.0 & df['BMI']<=25.0),'Normal')
```

```python
def change (bmi):
  if 18.0 <= bmi <= 25.0:
    return "Normal"
  return "Out Normal"

df['Normal Range'] = df['BMI'].apply(change)
df
```
