```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Generar datos sintéticos
np.random.seed(0)
horas_estudio = np.random.uniform(1, 10, 100)  # Horas de estudio
ruido = np.random.normal(0, 2, 100)  # Ruido
puntaje = 3 * horas_estudio + 5 + ruido  # Relación lineal + ruido
```

```python
# Visualizar los datos
plt.scatter(horas_estudio, puntaje, label='Datos')
plt.xlabel('Horas de Estudio')
plt.ylabel('Puntaje en el Examen')
plt.title('Relación entre Horas de Estudio y Puntaje en el Examen')
plt.legend()
plt.show()
```

```python
# Ajustar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(horas_estudio.reshape(-1, 1), puntaje)
```

```python
# Visualizar la línea de regresión
plt.scatter(horas_estudio, puntaje, label='Datos')
plt.plot(horas_estudio, modelo.predict(horas_estudio.reshape(-1, 1)), color='red', label='Regresión')
plt.xlabel('Horas de Estudio')
plt.ylabel('Puntaje en el Examen')
plt.title('Horas de Estudio vs Puntaje en el Examen')
plt.legend()
plt.show()
```

```python
# Obtener los coeficientes
coeficiente = modelo.coef_
intercepto = modelo.intercept_

print(f'Pendiente (m): {coeficiente[0]:.2f}')
print(f'Ordenada al origen (b): {intercepto:.2f}')
```

```python
# Evaluar el modelo
y_pred = modelo.predict(horas_estudio.reshape(-1, 1))
error = mean_squared_error(puntaje, y_pred)
print("Error cuadrático medio:", error)
```
