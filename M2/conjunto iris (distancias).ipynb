```python
import numpy as np
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path="/content/drive/MyDrive/Universidad/Cursos/Python Data Science/Datasets/M2/Iris.csv"
df = pd.read_csv(path)
df
```

```python
df_iris = df.drop(columns=['Id'])
df_iris
```

```python
prom_especie = df_iris.groupby('Species').mean()
prom_especie
```

```python
prom_especie.plot(kind='scatter', x='SepalWidthCm', y='SepalLengthCm', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
prom_especie.plot(kind='scatter', x='PetalWidthCm', y='PetalLengthCm', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
def Dis_Euclidiana(x1,x2,y1,y2):

  x = x2-x1
  y = y2-y1
  xy = (x)**2 + (y)**2
  distancia = np.sqrt(xy)

  return distancia
```

```python
#Distancia euclidiana setosa

y1 = float(prom_especie.iloc[0, 0])
y2 = float(prom_especie.iloc[0, 2])
x1 = float(prom_especie.iloc[0, 1])
x2 = float(prom_especie.iloc[0, 3])

y1,y2,x1,x2
```

```python
Setosa =Dis_Euclidiana(x1,x2,y1,y2)
```

```python
print("Distancia ecluidiana de sepalos Setosa - Versicolor: ",Setosa )
```

```python
#Distancia versicolor

y1 = float(prom_especie.iloc[1, 0])
y2 = float(prom_especie.iloc[1, 2])
x1 = float(prom_especie.iloc[1, 1])
x2 = float(prom_especie.iloc[1, 3])

y1,y2,x1,x2
```

```python
Versicolor = Dis_Euclidiana(x1,x2,y1,y2)
```

```python
print("Distancia ecluidiana de sepalos Versicolor - Virginica: ", Versicolor)
```

```python
#Distancia virginica

y1 = float(prom_especie.iloc[2, 0])
y2 = float(prom_especie.iloc[2, 2])
x1 = float(prom_especie.iloc[2, 1])
x2 = float(prom_especie.iloc[2, 3])

y1,y2,x1,x2
```

```python
Virginica = Dis_Euclidiana(x1,x2,y1,y2)
```

```python
print("Distancia ecluidiana de sepalos Setosa - Virginica: ", Virginica)
```

```python
# Create a list of the three variables
variables = [Virginica, Setosa, Versicolor]

# Create a bar chart
plt.bar(range(len(variables)), variables)

# Set the title and axis labels
plt.title("Comparison of Distances")
plt.xlabel("Species")
plt.ylabel("Distance")

# Set the x-axis tick labels
plt.xticks(range(len(variables)), ["Virginica", "Setosa", "Versicolor"])

# Show the plot
plt.show()
```

#DISTANCIA MANHATAN

```python
m = prom_especie.mean()
m
```

```python
def dis_Manhatan(x1,x2,y1,y2):
  x = abs(x2-x1)
  y = abs(y2-y1)
  distancia = x+y
  return distancia
```

```python
x1 = m['SepalWidthCm']
x2 = m['PetalWidthCm']
y1 = m['SepalLengthCm']
y2 = m['PetalLengthCm']
```

```python
dm = dis_Manhatan(x1,x2,y1,y2)
print("La distancia Manhatan entre A y B es: ",dm)
```

```python
#plt.figure(figureSize=(6,6))

plt.figure(figsize=(6,6))
plt.plot([x1,x2],[y1,y1], 'bo')
plt.plot([x2,x2],[y1,y2], 'bo')

plt.plot([x1,x2],[y1,y1], 'k-')
plt.plot([x2,x2],[y1,y2], 'k-')

plt.text(x1,y1,'A',ha='left')
plt.text(x2,y2,'B',ha='right')
y_mid = (y1+y2)/2
x_mid = (x1+x2)/2
plt.text(x_mid,y_mid,f'{dm}',ha='center')

plt.xlabel('Longitud Sépalo')
plt.ylabel('Longitud Pétalo')
plt.title('Distancia Manhatan A - B')

plt.grid(True)
plt.show()
```

#DISTANCIA EUCLIDIANA

```python
prom_especie.loc['Iris-setosa']
```

```python
set_m = np.array(prom_especie.loc['Iris-setosa'])
ver_m = np.array(prom_especie.loc['Iris-versicolor'])
vir_m = np.array(prom_especie.loc['Iris-virginica'])
```

```python
set_m
```

setosa - versicolor

```python
dif_set_ver = set_m - ver_m
dif_set_ver
```

```python
dE_set_ver = np.sqrt(np.sum(dif_set_ver**2))
print(f"La distancia euclidiana entre Setosa y Versicolor es: {dE_set_ver} centimetros")
```

```python
# Define the coordinates of the points
dif_set_ver = abs(dif_set_ver)
x1, y1 = dif_set_ver[1], dif_set_ver[0]
x2, y2 = dif_set_ver[3], dif_set_ver[2]

# Create the plot
plt.figure()

# Plot the points
plt.plot(x1, y1, 'bo')
plt.plot(x2, y2, 'bo')

# Draw a line connecting the points
plt.plot([x1, x2], [y1, y2], 'k-')

# Add text labels to the points
plt.text(x1, y1, 'Setosa', ha='right')
plt.text(x2, y2, 'Versicolor', ha='left')

# Set the axis labels and title
plt.xlabel('Sepalo (cm)')
plt.ylabel('Petalo (cm)')
plt.title('Distancia entre Setosa y Versicolor')

# Show the plot
plt.show()
```
