```python
import numpy as np
```

#Distancia euclidiana

```python
def Dis_Euclidiana(x1,x2,y1,y2):

  x = x2-x1
  y = y2-y1
  xy = (x)**2 + (y)**2
  distancia = np.sqrt(xy)

  return distancia
```

```python
x1 = 2
x2 = 6
y1= 6
y2= 8

d = Dis_Euclidiana(x1,x2,y1,y2)
d
```

#Distancia Manhattan

```python
def Dis_Manhattan(x1,x2,y1,y2):

  x = abs(x2-x1)
  y = abs(y2-y1)
  distancia = x + y

  return distancia
```

```python
x1 = 2
x2 = 6
y1= 6
y2= 8

d = Dis_Manhattan(x1,x2,y1,y2)
d
```

#Distancia de Chebyshev

```python
def Dis_Chebyshev(x1,x2,y1,y2):

  x = abs(x2-x1)
  y = abs(y2-y1)
  distancia = max(x,y)

  return distancia
```

```python
x1 = 2
x2 = 6
y1= 6
y2= 8

d = Dis_Chebyshev(x1,x2,y1,y2)
d
```

#Distancia de Canberra

```python
def Dis_Canberra(x1,x2,y1,y2):

  x = abs(x2-x1)
  y = abs(y2-y1)
  distancia = (x/(abs(x1)+abs(x2))) + (y/(abs(y1)+abs(y2)))

  return distancia
```

```python
x1 = 2
x2 = 6
y1= 6
y2= 8

d = Dis_Canberra(x1,x2,y1,y2)
d
```

#Distancia Coseno

```python
def Dis_Coseno(x1,x2,y1,y2):

  xy = np.dot(x1,y1) + np.dot(x2,y2)
  x = np.sqrt(np.dot(x1,x1) + np.dot(x2,x2))
  y = np.sqrt(np.dot(y1,y1) + np.dot(y2,y2))
  distancia = 1 - (xy/(x*y))

  return distancia
x1 = [1,1]
x2 = [1,0]
y1= [1,0]
y2= [1,1]

d = Dis_Coseno(x1,x2,y1,y2)
d
```

#Distancia de Jaccard

```python
def Dis_Jaccard(x1,x2,y1,y2):

  xy = len(set(x1).intersection(y1)) + len(set(x2).intersection(y2))
  x = len(set(x1).union(y1)) + len(set(x2).union(y2))
  distancia = 1 - (xy/x)

  return distancia
```

```python
x1 = [1,1]
x2 = [1,0]
y1= [1,0]
y2= [1,1]

d = Dis_Jaccard(x1,x2,y1,y2)
d
```
