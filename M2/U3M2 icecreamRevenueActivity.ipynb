```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
#from sklearn.metrics import mean_squared_error
```

```python
path='/content/drive/MyDrive/Universidad/Cursos/Python Data Science/Datasets/M2/IceCreamData.csv'
df = pd.read_csv(path)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,8))
sns.jointplot(x = 'Temperature', y= 'Revenue', data = df, color = 'green')
```

```python
x = df[['Temperature']]
y = df[['Revenue']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept = True)
```

```python
model.fit(X_train, y_train)
```

```python
yp = model.predict(X_test)
yp
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, model.predict(X_train), color = 'red')
plt.xlabel('Temperature')

plt.ylabel('Revenue')
plt.title('Revenue vs. Temperature')
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X_test, y_test)
plt.plot(X_test, model.predict(X_test), color = 'red')
plt.xlabel('Temperature')

plt.ylabel('Revenue [Naira]')
plt.title('Revenue vs. Temperature')
```

```python
acuracy = model.score(X_test, y_test)
acuracy
```
